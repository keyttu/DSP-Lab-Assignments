### ${\frac{1}{3^n}}u[n]$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
n= np.arange(-20,21)
```

```python
n[40]
```

```python
def signal(a):
    if(a>=0):
        return 1/(3**a)
    else:
        return 0
    
```

```python
y = np.zeros(41)
for i in range(41):
    y[i]= signal(n[i])
```

```python
plt.stem(n, y)
plt.xlabel(' Sequence n')
plt.ylabel('Ámplitude y(n)')
plt.show()
```

### $3 \cos \left[ \frac{\pi n}{4} + \frac{\pi}{3} \right]$

```python
def signal2(a):
    return 3*(math.cos(a*0.25*np.pi + np.pi*(1/3)))
```

```python
y1 = np.zeros(41)
for i in range(41):
    y1[i]= signal2(n[i])
```

```python
plt.stem(n, y1)
plt.xlabel(' Sequence n')
plt.ylabel('Ámplitude y(n)')
plt.show()
```

### $\frac{\sin (\pi n/6 ) }{\pi n/6}$

```python
def signal4(a):
    if(a==0):
        return 1
    else:
        return 6*(math.sin(np.pi*a*(1/6)))*(1/a)*(1/np.pi)
```

```python
y4 = np.zeros(41)
for i in range(41):
    y4[i]= signal4(n[i])
    
plt.stem(n, y4)
plt.xlabel(' Sequence n')
plt.ylabel('Ámplitude y4(n)')
plt.show()
```

```python
Ak = np.array([2,4,3,4])
wk = np.array([0, np.pi/20, 3*np.pi/40, 3*np.pi/20])
tetak = np.array([0, -np.pi/5, -np.pi/2, -np.pi/3])
```

```python
Ak[3]
```

```python
n = np.arange(-100,100)
sig = np.zeros(len(n))
for i in n:
    for j in range(4):
        sig[i+100] += Ak[j]*math.cos(wk[j]*i+tetak[j])
        
```

```python
plt.stem(n, sig)
plt.show()
```
