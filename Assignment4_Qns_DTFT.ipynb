Q1. Compute the DTFT (magnitude and phase) of the following (use <i>scipy.signal.freqz</i>). Plot from $\omega = −2\pi$ to $2\pi$. Observe the symmetries and relations between the spectra.

https://docs.scipy.org/doc/scipy-0.18.1/reference/generated/scipy.signal.freqz.html

<ol>
  <li>$r[n] = u[n] − u[n − 5]$  </li>
 <li>$r[n − 7]$  </li>
  <li>$r[n + 4]$ </li>
    <li>$r[−n]$ </li>
    <li>$(−1)^nr[n]$ </li>
</ol>

```python
import matplotlib.pyplot as plt
import numpy
import scipy.signal as sig

w = np.linspace(-2*np.pi, 2*np.pi, 1000)

# DTFT FORMULA

def DTFT(a, x, stp):
    X = 0
    n = np.arange(stp, stp+len(x))
    for i in n:
        X+= x[i-stp]*np.exp(-(1j)*a*i)
    return abs(X), np.angle(X)

x = np.ones(5)
stp = 0
X_abs = np.zeros(len(w))
X_ang = np.zeros(len(w))
for i in range(len(w)):
    X_abs[i], X_ang[i] = DTFT(w[i], x, stp) 
```

$ z[n]= \begin{cases}x[\frac{n}{2}]~~,~~~n~~even\\0~~,~~~~n~~~odd\end{cases}$

Q4. Consider the signals $x[n] = n(u[n]- u[n-4])$ and $y[n] = 0.9^n(u[n] - u[n-10])$. Find the convolution $z[n] = x[n] * y[n]$ of the signals. Plot $X(e^{j\omega})$, $Y(e^{j\omega})$ and $Z(e^{j\omega})$ (magnitude and phase separately).  


```python
##code
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

w = np.linspace(-2*np.pi, 2*np.pi, 1000)

def step(a):
    if(a<0):
        return 0
    else:
        return 1

n = np.arange(0,10)
x = np.zeros(len(n))
y = np.zeros(len(n))
for i in n:
    x[i] = (step(i) - step(i-4))*i
    y[i] = (step(i) - step(i-10))*(0.9)**i
    
plt.stem(n,x,'r')
plt.show()
plt.stem(n,y,'r')
plt.show()


stp = 0
X_abs_6 = np.zeros(len(w))
X_ang_6 = np.zeros(len(w))
for i in range(len(w)):
    X_abs_6[i], X_ang_6[i] = DTFT(w[i], x, stp) 
plt.plot(w, 20*np.log10(X_abs_6))
plt.show()
plt.plot(w, X_ang_6)
plt.show()

X_abs_7 = np.zeros(len(w))
X_ang_7 = np.zeros(len(w))
for i in range(len(w)):
    X_abs_7[i], X_ang_7[i] = DTFT(w[i], y, stp) 
plt.plot(w, 20*np.log10(X_abs_7))
plt.show()
plt.plot(w, X_ang_7)
plt.show()

Z_abs = X_abs_6*X_abs_7
Z_ang = X_ang_6 + X_ang_7

plt.plot(w, 20*np.log10(Z_abs))
plt.show()
plt.plot(w, Z_ang)
plt.show()

z_func = np.convolve(x,y)

X_abs_8 = np.zeros(len(w))
X_ang_8 = np.zeros(len(w))
for i in range(len(w)):
    X_abs_8[i], X_ang_8[i] = DTFT(w[i], z_func, stp) 
plt.plot(w, 20*np.log10(X_abs_8))
plt.show()
plt.plot(w, X_ang_8)
plt.show()
```

Q8. For the LTI systems described by the following difference equation, generate the frequency response. Obtain the pole zero plot. Comment on the type of response.

$y[n] + 1.3y[n - 1] + 1.29y[n - 2] + 0.52y[n - 3] + 0.16y[n - 4] = x[n] - 0.9x[n - 1]$

$y[n] - 1.3y[n - 1] + 1.29y[n - 2] - 0.52y[n - 3] + 0.16y[n - 4] = x[n] + 0.9x[n - 1]$


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig


def plot_zeros_poles(b,a):
    [z,p,k] = sig.tf2zpk(b,a)
    t3 = plt.plot(p.real, p.imag, 'rx')
    t4 = plt.plot(z.real, z.imag, 'b.')
    plt.setp(t3, markersize=12.0, markeredgewidth=3.0)
    plt.setp(t4, markersize=12.0, markeredgewidth=3.0)
    plt.show()
    
b = np.array([1,-0.9])
a = np.array([1,1.3,1.29,0.52])
n, y = sig.freqz(b,a)

plt.plot(n,y)
plt.show()
plot_zeros_poles(b,a)
b = np.array([1,0.9])
a = np.array([1,-1.3,1.29,-0.52])
n, y = sig.freqz(b,a)

plt.plot(n,y)
plt.show()
plot_zeros_poles(b,a)
```
